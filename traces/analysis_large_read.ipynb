```python
def get_trace_reference(path):
    
    file = open(path)  

    read_size1_count= 0
    
    read_size2_count=0
    
    read_size3_count=0
    
    read_size4_count=0
      
    page_size = 32  # 32 sector
    
    sub_page_size =8 # 8 sector  4KB
 
    capacity = 32 * 1024 * 1024  # 512G SSD = 32 * 1024* 1024*16KB   
    
    file = open(path) 
    while 1:
        lines = file.readlines(100000)
        if not lines:
            break
        for line in lines:
            tmp = line.split()
            #print(line)
            if(int(tmp[4]) == 1):  # read
                
                first_lpn = int(int(tmp[2]) / page_size) % capacity
                
                
                if( int(  int(tmp[2])- page_size*first_lpn)!=0):
                                        
                    if( int(int(tmp[2])- page_size*first_lpn)<1*sub_page_size):
                        
                        read_size4_count+=1
                    
                    elif( int(int(tmp[2])- page_size*first_lpn)<2*sub_page_size):
                        
                        read_size3_count+=1

                    elif( int(int(tmp[2])- page_size*first_lpn)<3*sub_page_size):
                        
                        read_size2_count+=1
                        
                    elif(int(int(tmp[2])- page_size*first_lpn)<4*sub_page_size):
                        
                        read_size1_count+=1                       
                        
                    first_lpn+=1
                        

                while(first_lpn <= int( (int(tmp[2]) + int(tmp[3])) / page_size) % capacity):

                    read_size4_count+=1
                    
                    first_lpn = first_lpn + 1
                    
                    
                if( int( (int(tmp[2]) + int(tmp[3])) % page_size)  !=0):
                    
                    #read_size4_count-=1
                    
                    if(int( (int(tmp[2]) + int(tmp[3])) % page_size) <=1*sub_page_size):
                        
                        read_size1_count+=1
                    
                    elif(int( (int(tmp[2]) + int(tmp[3])) % page_size) <=2*sub_page_size):
                        
                        read_size2_count+=1

                    elif(int( (int(tmp[2]) + int(tmp[3])) % page_size) <=3*sub_page_size):
                        
                        read_size3_count+=1
                        
                    elif(int( (int(tmp[2]) + int(tmp[3])) % page_size) <=4*sub_page_size):
                        
                        read_size4_count+=1                        
                        
                        
                        
            if(int(tmp[4]) == 0):  # write
                
                first_lpn = int(int(tmp[2]) / page_size) % capacity
                
                if( (int(tmp[2])  - page_size*first_lpn)!=0):                   
                    
                    if( (int(tmp[2])  - page_size*first_lpn)<=1*sub_page_size):  
                        
                        read_size1_count+=1
                    
                    elif( (int(tmp[2])  - page_size*first_lpn)<=2*sub_page_size):
                        
                        read_size2_count+=1

                    elif( (int(tmp[2])  - page_size*first_lpn)<=3*sub_page_size):
                        
                        read_size3_count+=1
                    elif( (int(tmp[2])  - page_size*first_lpn)<4*sub_page_size):
                        read_size3_count+=1
                                                
                        
                if(   ( (  int(tmp[2]) + int(tmp[3]) ) / page_size )!=0):
                                  
                    if((page_size - int(tmp[2]) % page_size)<=1):
                        
                        read_size1_count+=1
                    
                    elif((page_size - int(tmp[2]) % page_size)<=2):
                        
                        read_size2_count+=1

                    elif((page_size - int(tmp[2]) % page_size)<=3):
                        
                        read_size3_count+=1
                        
                    elif((page_size - int(tmp[2]) % page_size)<=4):
                        
                        read_size4_count+=1                        
                        
                        
         
    file.close()
    
    readsum=read_size1_count+read_size2_count+read_size3_count+read_size4_count
    
    print("%s %.2f %.2f %.2f  %.2f"%(path,read_size1_count/readsum,read_size2_count/readsum,read_size3_count/readsum,read_size4_count/readsum))

tracename = ["HM_0", "HM_1", "PRN_0", "PRN_1", "PROJ_1", "PROJ_3", "PROJ_4", "PRXY_0", "RSRCH", "SRC1_2", "SRC2_0", "SRC2_2", "STG_0", "USR_0", "WDEV_0"]

for item in tracename:
    
    get_trace_reference(item)

```
